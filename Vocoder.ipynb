## Spectral envelope parametrization & recovering

```python
def decompose(x, fs, period=5.0):
    """using WORLD, waveform = F0 + spectral envelope (sp) + aperiodicity (ap)
    """
    f0, timeaxis = pyworld.harvest(x, fs, frame_period=period)
    sp = pyworld.cheaptrick(x, f0, timeaxis, fs)
    ap = pyworld.d4c(x, f0, timeaxis, fs)
    return f0, timeaxis, sp, ap
```

```python
def vis(sp_param, approx_sp, x, y, top_title):
    """Visualization of compressed spectral parameter, recovered spectral envelope, reconstructed speech wavevform
    """
    figure(figsize=(16,14))
    subplots_adjust(hspace=0.4)
    subplot(3,1,1)
    librosa.display.specshow(sp_param.T, sr=sampling_rate, hop_length=hop_length, x_axis="time", y_axis="linear")
    title(top_title)
    subplot(3,1,2)
    librosa.display.specshow(10*log(approx_sp).T, sr=sampling_rate, hop_length=hop_length, x_axis="time", y_axis="linear")
    title("20log|H(w)|")
    subplot(3,1,3)
    plot(y, "r-+", label="reconstructed speech signal")
    plot(x, label="original speech signal")
    xlabel("sample")
    legend(fontsize=14)
```

```python
def mcepalpha(fs):
    """Determine frequency warping parameter from fs (simplest implementation)
    See https://bitbucket.org/happyalu/mcep_alpha_calc/ for details.
    """
    if fs == 16000:
        return 0.41
    elif fs == 22050:
        return 0.455
    elif fs == 44100:
        return 0.544
    elif fs == 48000:
        return 0.554
    else:
        raise NotImplementedError
```

## Recovering from coded spectral envelope by WORLD

```python
dim = 60
coded_sp = pyworld.code_spectral_envelope(sp, sampling_rate, dim)
decoded_sp = pyworld.decode_spectral_envelope(coded_sp, sampling_rate, fftlen)
y = pyworld.synthesize(f0, decoded_sp, ap, sampling_rate, frame_period)

nmse = np.linalg.norm(np.log(sp) - np.log(decoded_sp)) / np.linalg.norm(np.log(sp))
print("NMSE: {}".format(nmse))

vis(coded_sp, decoded_sp, x, y, "Mel-cepstrum")
Audio(y, rate=sampling_rate)
```

## Recovering from mel-cepstrum using SPTK

```python
dim = 60
alpha = mcepalpha(sampling_rate)
mc = np.apply_along_axis(pysptk.sp2mc, 1, sp, dim-1, alpha)
approximate_sp = np.apply_along_axis(pysptk.mc2sp, 1, mc, alpha, fftlen)
y = pyworld.synthesize(f0, approximate_sp, ap, sampling_rate, frame_period)

nmse = np.linalg.norm(np.log(sp) - np.log(approximate_sp)) / np.linalg.norm(np.log(sp))
print("NMSE: {}".format(nmse))

vis(mc, approximate_sp, x, y, "Coded spectral envelope by pysptk")
Audio(y, rate=sampling_rate)
```

## Estimate spectral envelop and parameters

```python
# plotting utility
def pplot(sp, envelope, title="no title"):
    plot(sp, "b-", linewidth=2.0, label="Original log spectrum 20log|X(w)|")
    plot(20.0/np.log(10)*envelope, "r-", linewidth=3.0, label=title)
    xlim(0, len(sp))
    xlabel("frequency bin")
    ylabel("log amplitude")
    legend(prop={'size': 20})
```

```python
# Compute spectrum 20log|X(w)| for a windowed signal
sp = 20*np.log10(np.abs(np.fft.rfft(xw)))
```

```python
mgc = pysptk.mgcep(xw, 20, 0.0, 0.0)
pplot(sp, pysptk.mgc2sp(mgc, 0.0, 0.0, frame_length).real, title="Lineaer frequency cepstrum based envelope")
```

```python
mgc = pysptk.mcep(xw, 20, 0.41)
pplot(sp, pysptk.mgc2sp(mgc, 0.41, 0.0, frame_length).real, title="Mel-cepstrum based envelope")
```

```python
mgc = pysptk.mgcep(xw, 20, 0.0, -1.0)
pplot(sp, pysptk.mgc2sp(mgc, 0.0, -1.0, frame_length).real, title="LPC cepstrum based envelope")
```

```python
mgc = pysptk.mgcep(xw, 20, 0.41, -1.0)
pplot(sp, pysptk.mgc2sp(mgc, 0.41, -1.0, frame_length).real, title="Warped LPC cepstrum based envelope")
```

```python
mgc = pysptk.gcep(xw, 20, -0.35)
pplot(sp, pysptk.mgc2sp(mgc, 0.0, -0.35, frame_length).real, title="Generalized cepstrum based envelope")
```

```python
mgc = pysptk.mgcep(xw, 20, 0.41, -0.35)
pplot(sp, pysptk.mgc2sp(mgc, 0.41, -0.35, frame_length).real, title="Mel-generalized cepstrum based envelope")
```

## F0 estimation using two SPTK-SWIPE, SPTK-RAPT and WORLD.

```python
def world_decompose(wav, fs, frame_period = 5.0):

    # Decompose speech signal into f0, spectral envelope and aperiodicity using WORLD
    wav = wav.astype(np.float64)
    f0, timeaxis = pyworld.harvest(wav, fs, frame_period = frame_period, f0_floor = 71.0, f0_ceil = 800.0)
    sp = pyworld.cheaptrick(wav, f0, timeaxis, fs)
    ap = pyworld.d4c(wav, f0, timeaxis, fs)

    return f0, timeaxis, sp, ap
```

```python
f0_swipe = pysptk.swipe(x.astype(np.float64), fs=sampling_rate, hopsize=80, min=60, max=200, otype="f0")
plot(f0_swipe, linewidth=3, label="F0 trajectory estimated by SWIPE'")

f0 = pysptk.rapt(x.astype(np.float32), fs=sampling_rate, hopsize=80, min=60, max=200, otype="f0")
plot(f0, linewidth=3, label="F0 trajectory estimated by RAPT")

f0, timeaxis, sp, ap = world_decompose(wav = x, fs = 16000, frame_period = 5.0)
plot(f0, linewidth=3, label="F0 trajectory estimated by WORLD")

plt.xlabel('frame')
xlim(0, len(f0_swipe))
legend(prop={'size': 18})
```
